# `SweetViz` Example

```python
import pandas as pd
import sweetviz as sv
```

## Load Sample Data

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

## Generate EDA Report

```python
my_report = sv.analyze(
    source = iris,
)
```

## Show EDA Report

```python
# Write the report out to a HTML file named "sweetviz-iris.html"
# my_report.show_html(
#     filepath="sweetviz-iris.html"
# )

my_report.show_notebook()
```

## Compare Two Subsets

```python
my_report_compare = sv.compare_intra(
    source_df = iris,
    condition_series = iris['species'] == 'versicolor',
    names = ['Versicolor', 'Non-versicolor']
)
```

```python
my_report_compare.show_notebook()
```
